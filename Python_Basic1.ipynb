```python
""" 
15) Data Types : 

In Python , the following inbuilt Data Type exist(14 inbuilt dataType) 
Primitive Types: Basic data types such as integers (int), floating-point numbers (float), and strings (str) are all objects in Python. 
For example, the integer 5 is an instance of the int class.

14) INbuilt - Data Types 
int , float , complex , bool , str , bytes , bytearray , range , list , tuple , set , frozenset, dict , None .

->there is no char datatype

In Python, the term "object" encompasses all data types because everything in Python is an object. 
This includes primitive data types, complex data types, functions, classes, and more.
Here’s a breakdown to clarify how objects and data types relate in Python:

Data types themselves are classes (objects) that define how values are stored and manipulated, 
and each instance of these data types is also an object. 
Therefore, the concept of "object" in Python is foundational and applies to all types of data 
and functionality within the language.

Data Types as Objects:



Composite Types: More complex data types like lists (list), dictionaries (dict), and tuples (tuple) are also objects. 
For instance, [1, 2, 3] is an instance of the list class.

NOTE :
Objects Have Identity, Type, and Value.

=>Identity: Each object has a unique identity (can be obtained using id()) and The value returned by id() is an integer.

In Python, the id() function returns the unique identifier for an object. 
id() provides a unique identifier for the object, which is unique among all objects during the lifetime of that object.
It is generally implemented as the memory address where the object is stored.


=>Type: Each object has a type (can be obtained using type()).

=>Value: Each object holds some value or data.


Usage:
The id() function is used to compare objects and determine if they are the same object in memory. 
This can be useful for understanding object identity and verifying whether two references point to the same object.
a = 10
print(id(a))  # Unique identifier for the integer object
print(type(a))  # Output: <class 'int'>

The id() value is valid for the duration of the object’s existence. Once the object is deleted or goes out of scope, its id() value may be reused for new objects.


"""
```
